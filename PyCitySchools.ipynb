# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually underperformed compared to schools with smaller budgets.


* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).


* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculating desired values

# 1. Total number of schools
school_total = len(school_data_complete['school_name'].value_counts())
```

```python
# 2. Total number of students
student_total = school_data_complete['Student ID'].count()
```

```python
# 3. Total Budget
budget_total = sum(school_data_complete['budget'].unique())
```

```python
# 4. Average math score
math_score_avg = school_data_complete['math_score'].mean()
```

```python
# 5. Average reading score
reading_score_avg = school_data_complete['reading_score'].mean()
```

```python
# 6. Overall average passing rate
pass_rate = (math_score_avg + reading_score_avg) / 2
```

```python
# 7. Percentage of students with a passing math score (70 or greater)
student_math_70 = school_data_complete[school_data_complete['math_score'] >= 70]
math_percent_70 = len(student_math_70) / student_total * 100
```

```python
# 8. Percentage of students with a passing reading score (70 or greater)
student_reading_70 = school_data_complete[school_data_complete['reading_score'] >= 70]
reading_percent_70 = len(student_reading_70) / student_total * 100
```

```python
# Creating new dataframe containing all values

district_summary = {
    'Total Number of Schools': [school_total], 
    'Total Number of students': ['{:,}'.format(student_total)], 
    'Total Budget': ['${:,}'.format(budget_total)], 
    'Average Math Score': [round(math_score_avg, 1)], 
    'Average Reading Score': [round(reading_score_avg, 1)], 
    'Overall Average Score': [round(pass_rate, 2)], 
    'Students Passing Math': ['%{:,}'.format(round(math_percent_70, 1))], 
    'Students Passing Reading': ['%{:,}'.format(round(reading_percent_70, 1))]
}

district_summary_df = pd.DataFrame(district_summary)
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
# 1. Grouping by school names - School name used as index
grouped_schools = school_data_complete.groupby(['school_name'])
```

```python
# 2. Finding school types
school_type = grouped_schools['type'].unique()
```

```python
# 3. Total students per school
school_students = grouped_schools['student_name'].size()
```

```python
# 4. Total budget per school
school_budget = grouped_schools['budget'].unique()
```

```python
# 5. Budget per student
student_budget = school_budget / school_students
```

```python
# 6. School average math score
school_math_avg = grouped_schools['math_score'].mean()
```

```python
# 7. School average reading score
school_reading_avg = grouped_schools['reading_score'].mean()
```

```python
# 8. Percentage of students with a passing math score (70 or greater)
school_math_70 = school_data_complete[school_data_complete['math_score']>=70].groupby(['school_name']).size()
perc_school_math_70 = school_math_70 / school_students * 100
```

```python
# 9. Percentage of students with a passing reading score (70 or greater)
school_reading_70 = school_data_complete[school_data_complete['reading_score']>=70].groupby(['school_name']).size()
perc_school_reading_70 = school_reading_70 / school_students * 100
```

```python
# 10. Overall passing percentage
perc_school_passing = (perc_school_math_70 + perc_school_reading_70) / 2
```

```python
# Creating a raw dataframe to hold the above results
# raw data frame will be used later for more analysis

school_summary = {'School Type': school_type, 
                  'Total Students': school_students, 
                  'Total School Budget': school_budget, 
                  'Per Student Budget': student_budget, 
                  'Average Math Score': round(school_math_avg, 1), 
                  'Average Reading Score': round(school_reading_avg, 1), 
                  '% Passing Math': round(perc_school_math_70, 1), 
                  '% Passing Reading': round(perc_school_reading_70, 1), 
                  '% Overall Passing': round(perc_school_passing, 1)}


raw_school_summary_df = pd.DataFrame(school_summary)
```

```python
# Removing brackets in School Type and School/Student Budgets
formatted_school_summary = raw_school_summary_df.copy()
formatted_school_summary['Per Student Budget'] = formatted_school_summary['Per Student Budget'].str[0]
formatted_school_summary['Total School Budget'] = formatted_school_summary['Total School Budget'].str[0]
formatted_school_summary['School Type'] = formatted_school_summary['School Type'].str[0]

# Formating numbers
formatted_school_summary['Per Student Budget'] = formatted_school_summary['Per Student Budget'].map('${:,.1f}'.format)
formatted_school_summary['Total School Budget'] = formatted_school_summary['Total School Budget'].map('${:,.1f}'.format)
formatted_school_summary.index.name = "School Name"
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["< $585", "$585-615", "$615-645", "$645-675"]
```

```python
# Binning the values based off school budget per student
raw_school_summary_df['binned'] = pd.cut(raw_school_summary_df['Per Student Budget'], spending_bins, labels = group_names)

school_summary_binned = raw_school_summary_df.groupby('binned')

# Calculating average math scores in each bin
avg_binned_math = school_summary_binned['Average Math Score'].mean()

# Calculating average reading scores in each bin
avg_binned_reading = school_summary_binned['Average Reading Score'].mean()

# Calculating average math pass scores in each bin
perc_binned_pass_math = school_summary_binned['% Passing Math'].mean()

# Calculating average reading pass scores in each bin
perc_binned_pass_reading = school_summary_binned['% Passing Reading'].mean()

# Creating new dataframe to contain above values
school_summary_binned_values = {
    'Average Math Score': round(avg_binned_math, 1), 
    'Average Reading Score': round(avg_binned_reading, 1), 
    '% Passing Math': round(perc_binned_pass_math, 1),
    '% Passing Reading': round(perc_binned_pass_reading, 1)
}

school_summary_binned_values_df = pd.DataFrame(school_summary_binned_values)

# Renaming index 
school_summary_binned_values_df.index.name = 'Spending Range per Student'

school_summary_binned_values_df
```
